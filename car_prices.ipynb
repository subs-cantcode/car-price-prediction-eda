#### **Comparing numerical columns with target**

````python
# important visualisation libraries

import matplotlib.pyplot as plt
import seaborn as sns
````

````python
#comparing odometer with selling price (scatterplot)

sns.scatterplot(x='odometer', y = 'sellingprice', data=df)
plt.title("comparing odometer with selling price")
plt.show()

#insights:
#less the odometer, more is the selling price.
````

````python
# comparing year with selling price
# better to use scatterplot since year and sellingprice are numerical,
# and they're best if you're finding the relationship, overall trend or correlation between features and target

sns.scatterplot(x= 'year', y='sellingprice', data = df)
plt.title("Selling price by year")
plt.show()

#insights: 
# Selling prices generally increase for newer car models.
# Cars manufactured after 2005 show significantly higher price ranges, with the highest selling prices concentrated around 2014 models.
````

````python
# comparing mmr with sellingprice
#mmr, manheim market report, information about how much it is likely to sold for 

sns.scatterplot(x= 'mmr', y='sellingprice', data = df)
plt.title("Selling price by mmr")
plt.show()

#insights: 
# mmr close and strongly related to sellingprice 
# as the mmr increases, the selling price also increases
# most vehicles are sold close to their estimated market value
````

#### **Comparing categorical columns with target**

````python
#this is to compare most common body types based on average selling price, for the first 10 entries from the top
most_common_body_types = df.groupby('body')['sellingprice'].mean().sort_values(ascending=False).head(10)

#sns.barplot(x='body', y='sellingprice',  data='top_10_cars')
most_common_body_types.plot(kind='bar')
plt.title("Top 10 most common body types by Average Selling price")
plt.xlabel("Body Types")
plt.ylabel("Selling Price")
plt.show()

#insights:
#granturismo convertible which is the most recognised body type, has average 80000 on it's selling price, which indicates it's impact on upcoming car manufacturing models in the future market.
````

````python
# Do vehicles with automatic transmission have higher average selling prices than manual vehicles?

#the error: when i checked the highest number of occurences in transmission types, i found out incorrect categories inside the dataset ('sedan', 'Sedan'). These values represent vehicle body types rather than valid transmission categories.

#so i filtered them out to disply only automatic and manual occurences using .isin()

df = df[df['transmission'].isin(['automatic', 'manual'])] #finding how many times the different transmission types have occured

transmission_types= df['transmission'].value_counts()
transmission_types

compare_sp_trans = df.groupby('transmission')['sellingprice'].mean()

compare_sp_trans.plot(kind='bar')
plt.title("Average selling price on different Transmission types")
plt.xlabel("Transmission types")
plt.ylabel("Selling price")
plt.show()

#insights: 
# the vehicles with automatic transmission types have higher average selling price compared to manual types. 
````

````python
# we need to understand the average selling price influence from different car manufacturers or car brands

df['make'].value_counts()

top_car_brands = df.groupby('make')['sellingprice'].mean().sort_values(ascending=False).head(10)

top_car_brands.plot(kind='bar')
plt.title("Average Selling price of top 10 Different car brands")
plt.xlabel("Car brands")
plt.ylabel("Selling price")
plt.show()

# Economy and discontinued brands such as Daewoo, Geo, and Plymouth have the lowest average selling prices, 
# indicating lower market demand and resale value compared with premium manufacturers.
````
